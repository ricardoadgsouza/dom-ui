```python
import os
import json
import pandas as pd

pasta_json = "data"
saida_parquet = "dados/atos.parquet"
os.makedirs("dados", exist_ok=True)

records = []

for nome in os.listdir(pasta_json):
    if nome.endswith(".json"):
        caminho = os.path.join(pasta_json, nome)
        with open(caminho, encoding="utf-8") as f:
            dados = json.load(f)

        for ed in dados.get("edicoes_ordinarias_exclusivas", []):
            if not isinstance(ed, dict):
                continue
            if ed.get("municipio") != "Florianópolis":
                continue
            for ato in ed.get("atos", []):
                rec = ato.copy()
                rec["data_edicao"] = ed.get("data")
                rec["url_edicao"] = ed.get("url")
                records.append(rec)

df = pd.DataFrame(records)
df.to_parquet(saida_parquet, index=False)
print(f"✅ Parquet salvo com {len(df)} registros.")
```
